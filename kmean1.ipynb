```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv("student-por.csv")
data.head()
```

```python
plt.scatter(data["G1"],data["G2"])
```

```python
scater = MinMaxScaler()
```

```python
scater.fit(data[["G1"]])
data["G1"] = scater.transform(data[["G1"]])
scater.fit(data[["G2"]])
data["G2"] = scater.transform(data[["G2"]])

data.head()
```

```python
km = KMeans(n_clusters=3)
y_predict=km.fit_predict(data[["G1","G2"]])
y_predict
```

```python
data["cluster"]= y_predict
data.head()
```

```python
data1=data[data["cluster"]==0]
data2=data[data["cluster"]==1]

data3=data[data["cluster"]==2]
```

```python
plt.scatter(data1["G1"],data1["G2"],color="red")
plt.scatter(data2["G1"],data2["G2"],color="blue")
plt.scatter(data3["G1"],data3["G2"],color="yellow")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="pink",marker="*")
```

```python
sse=[]
for i in range(2,20):
    km = KMeans(n_clusters=i)
    km.fit(data[["G1","G2"]])
    sse.append(km.inertia_)
   
    
    
```

```python
plt.plot(range(2,20),sse)
```
